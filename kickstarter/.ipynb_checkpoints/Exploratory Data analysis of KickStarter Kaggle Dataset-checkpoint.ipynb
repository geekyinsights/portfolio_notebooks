Question: What countries are backers coming from? Are more backers coming from outside of the US or inside of the US? 

Answer: Most projects were funded in the US.

Method: Compared 'usd_pledged' and 'usd_real_pledged' and calculated the percent to the total contribution.


```python
df['backers'].groupby(df['country']).count()
```

```python
df['diff'] = df['usd_pledged_real'] - df['usd pledged']

test = df[df['diff'] >=1]

usd_pledged_in_us = test[df['country'] == 'US']

usd_pledged_in_us[usd_pledged_in_us['state'] == 'successful']
```

```python
df['diff'] = (df['usd_pledged_real'] / df['usd pledged']  ) * 100
test = df[df['diff']  < 100]

usd_pledged_in_us = test[df['country'] == 'US']

usd_pledged_in_us[usd_pledged_in_us['state'] == 'successful']
```
